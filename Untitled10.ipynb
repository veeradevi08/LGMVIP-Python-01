```python
import moviepy.editor as mp
from google.colab import files
from IPython.display import Image, display
```

```python
def create_gif(image_files, output_file, duration):
    # Load images into a list of moviepy.ImageClip
    clips = [mp.ImageClip(img).set_duration(duration) for img in image_files]

    # Concatenate all the ImageClips
    video_clip = mp.concatenate_videoclips(clips, method="compose")

    # Write the result to a GIF file
    video_clip.write_gif(output_file, fps=10)
```

```python
uploaded = files.upload()
```

```python
# Get the filenames of the uploaded images
image_files = list(uploaded.keys())

# Define the output file name
output_file = "output.gif"

# Set the duration for each image in the GIF (in seconds)
duration = 1  # Adjust as needed

# Create the GIF
create_gif(image_files, output_file, duration)

# Display the created GIF
display(Image(filename=output_file))
```
